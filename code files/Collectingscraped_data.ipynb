Using SerpApi to extract the data from walmart particularly  Skateboard product reviews.

Link to SerpAPi:  https://serpapi.com/


```python
from serpapi import GoogleSearch

params = {
  "engine": "walmart",
  "query": "Skateboard",
  "api_key": "ad944e728e7267dbfed3623b663cd23686ac14f1a9a9ccf42b1ca1947c4270ed"
}

search = GoogleSearch(params)
results = search.get_dict()
organic_results = results["organic_results"]
```

```python
len(organic_results)
```

```python
prod_id_lst = []

for i in range(len(organic_results)):
    prod_id_lst.append(organic_results[i]['us_item_id'])

print(prod_id_lst)
```

Collecting the reviews based on different product IDs

```python
from serpapi import GoogleSearch

reviewlst = []

for i, product_id in enumerate(prod_id_lst):
    print(f"Processing product {i+1}/{len(prod_id_lst)}")

    params = {
        "engine": "walmart_product_reviews",
        "product_id": product_id,
        "api_key": "ad944e728e7267dbfed3623b663cd23686ac14f1a9a9ccf42b1ca1947c4270ed"
    }

    search = GoogleSearch(params)
    results = search.get_dict()

    # Ensure there is no error and reviews exist
    if 'error' not in results and 'reviews' in results:
        for review in results['reviews']:
            # Only add review if 'text' exists
            if 'text' in review:
                reviewlst.append({
                    'id': review.get('position', None),
                    'review': review['text'],
                    'rating': review.get('rating', None)
                })

print(f"Total reviews collected: {len(reviewlst)}")
```
