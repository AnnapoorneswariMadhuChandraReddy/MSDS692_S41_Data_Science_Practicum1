Loading data from publicly available resources.

Dataset for YELP are retrieved from: https://business.yelp.com/data/resources/open-dataset/

```python
import pandas as pd
import tarfile
import json

tar_path = "yelp_dataset.tar"

rows = []
with tarfile.open(tar_path, "r:*") as tar:
    f = tar.extractfile("yelp_academic_dataset_review.json")

    for i, line in enumerate(f):
        rows.append(json.loads(line))
        if i == 99999:   # take first 1000 reviews
            break

df = pd.DataFrame(rows)
print(df.head())
```

The kindle reviews Data is retrived from: https://www.kaggle.com/datasets/bharadwaj6/kindle-reviews


```python
import pandas as pd

df_am = pd.read_csv("kindle_reviews.csv")
```

```python
df_am
```

```python

df = df[["text", "stars"]].rename(columns={"text": "review", "stars": "rating"})
df_am = df_am[["reviewText", "overall"]].rename(columns={"reviewText": "review", "overall": "rating"})
```

Getting the data scrapped from walmart using serpAPi

```python
import pandas as pd

df_wl = pd.read_csv("walmart_reviews.csv")
```

```python
df_wl
```

## Combining all data available and saving it.

```python
df_final = pd.concat(
    [df[['review', 'rating']], df_am[['review', 'rating']],df_wl[['review', 'rating']] ],
    ignore_index=True
)
df_final
```

```python
df_final.to_csv("df_complete_download.csv", index=False)

# df_human = df_human.sample(n=205760, random_state=42)
```
